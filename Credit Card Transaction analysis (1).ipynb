```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trans_data = pd.read_csv("D:\credit_card.csv")
trans_data.head()
```

```python
customer_data = pd.read_csv("C:/Users/pratiksha/Downloads/customer(1).csv")
customer_data.head()
```

```python
trans_data.isnull().sum()
```

### Create new column 'Revenue'

```python
trans_data['Revenue'] = trans_data.Annual_Fees+trans_data.Total_Trans_Amt+trans_data.Interest_Earned
trans_data.head()
```

### Create new column Age_Group in customer data

```python
def custom_age(age):
    if age < 30:
        return '20-30'
    elif age <= 40:  
        return '30-40'
    elif age <= 50: 
        return '40-50'
    elif age <= 60:  
        return '50-60'
    elif age >= 60:
        return '60+'
    else:
        return 'unknown'

    
customer_data['Age_Group'] = customer_data['Customer_Age'].apply(custom_age)
customer_data.head()
```

###  Create new column Income_Group in customer data


```python
def income(income):
    if income < 35000:
        return 'Low'
    elif income <= 70000:
        return 'Medium'
    elif income >= 70000:
        return 'High'
    else:
        'unknown'
customer_data['Income_Group'] = customer_data['Income'].apply(income)
customer_data.head()
```

### Revenue by Card Category

```python
Revenue_by_Card_Category = trans_data.groupby('Card_Category')['Revenue'].sum()
plt.figure(figsize=(7,5))

sns.barplot(x = 'Card_Category', y='Revenue', data = Revenue_by_Card_Category.reset_index())

plt.xticks(rotation=90, fontsize=14)
plt.title('Revenue by Card Category')
plt.xlabel('Card Category')
plt.ylabel('Revenue')
plt.show()
```

### Revenue by Expenditure Type

```python
Revenue_by_Expenditure_Type = trans_data.groupby('Exp Type')['Revenue'].sum().reset_index()
plt.figure(figsize=(7,5))

barplot = sns.barplot(x = 'Exp Type', y='Revenue', data = Revenue_by_Expenditure_Type.reset_index(),color = 'mediumpurple')

plt.xticks(rotation=90, fontsize=14)
plt.title('Revenue by Expenditure Type')
plt.xlabel('Expenditure Type')
plt.ylabel('Revenue')
plt.show()
```

### Proportion of Revenue By Chip

```python
Revenue_By_Chip = trans_data.groupby('Use Chip')['Revenue'].sum().reset_index()

plt.figure(figsize=(10,5))
plt.pie(Revenue_By_Chip['Revenue'], labels = Revenue_By_Chip['Use Chip'],autopct = '%1.1f%%')
plt.title('Proportion of Revenue By Chip')
plt.show()
```

### Revenue by Quarter and Use Chip

```python

# Plotting with seaborn
plt.figure(figsize=(10, 6))
barplot= sns.barplot(data=trans_data, x='Qtr', y='Revenue', hue='Use Chip')

plt.xlabel('Quarter')
plt.ylabel('Revenue')
plt.title('Revenue by Quarter and Use Chip')
plt.xticks(rotation=45)
plt.legend(title='Use Chip')
plt.tight_layout()

plt.show()
```

### Revenue by Customer Job

```python
import pandas as pd

merged_df = pd.merge(trans_data, customer_data, on='Client_Num')

Revenue_by_Customer_Job = merged_df.groupby('Customer_Job')['Revenue'].sum()
plt.figure(figsize=(7,5))

sns.barplot(x = 'Customer_Job', y='Revenue', data = Revenue_by_Customer_Job.reset_index())

plt.xticks(rotation=90, fontsize=14)
plt.title('Revenue by Customer Job')
plt.xlabel('Customer Job')
plt.ylabel('Revenue')
plt.show()
```
